```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy
```

```python
plt.style.use('fivethirtyeight')
```

```python
para=sns.load_dataset("tips")
```

```python
para.head()
```

# Plot a Scatter plot between total_bill and tip

```python
sns.relplot(x='total_bill', y='tip', kind='scatter', data=para)
```

# The hue parameter

```python
sns.relplot(x='total_bill', y='tip', hue='smoker', data=para)
```

# The style Parameter

```python
sns.relplot(x='total_bill', y='tip', hue='smoker', style='sex', data=para)
```

# The size Parameter

```python
sns.relplot(x='total_bill', y='tip', hue='smoker', size='size', sizes=(13,200), style='sex', data=para)
```

# Using Scatterplot function

```python
sns.scatterplot(x='total_bill', y='tip', hue='smoker', style='sex', data=para)
```

# -> Plotting with Categorical Data

# Categorical Scatterplots

# a.Stripplot

```python
sns.catplot(x='day', y='tip', kind='strip',jitter=3, data=para)
```

# b.Swarmplot

```python
sns.catplot(x='day', y='tip', kind='swarm', data=para)
```

# --> Categorical Distribution plots

# a. Boxplot

```python
sns.boxplot(para['total_bill'])
```

```python
sns.catplot(x='day', y='total_bill', kind='box', data=para)
```

```python
sns.catplot(x='day', y='total_bill', hue='sex', kind='box', data=para)
```

# b. Violinplot

```python
sns.violinplot(para['tip'])
```

```python
sns.catplot(x='day', y='total_bill', kind='violin', hue='sex',split=True, data=para)
```

# --> Categorical Estimate Plots

# a.Barplot

```python
sns.catplot(x='smoker', y='total_bill', kind='bar',estimator=np.std, data=para)
```

```python
sns.catplot(x='smoker', y='total_bill', kind='bar', hue='sex', data=para)
```

# b. Countplot

```python
sns.catplot(x='sex', kind='count', hue='smoker', data=para)
```

```python
datas=sns.load_dataset('flights')
```

```python
datas.head()
```

```python
x=datas.pivot_table(index='year', columns='month', values='passengers', aggfunc='sum')
```

```python
sns.heatmap(x, cbar=False)
```

```python
sns.heatmap(x, cbar=False, linewidth=1)
```

```python
sns.heatmap(x, cbar=False, linewidth=1, annot=True, fmt='d', cmap='winter')
```

# b.Clustermap

```python
import scipy
sns.clustermap(x)
```

# Jointplot

```python
sns.jointplot(x='total_bill', y='tip', data=para)
```

```python
sns.jointplot(x='total_bill', y='tip', kind='hex', data=para)
```

```python
sns.jointplot(x='total_bill', y='tip', kind='kde', data=para)
```

```python
sns.jointplot(x='total_bill', y='tip', kind='reg', data=para)
```

```python
sns.pairplot(datas, hue='month')
```
